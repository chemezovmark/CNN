```python
import tensorflow as tf
from keras import backend as K
from keras.applications.vgg16 import VGG16
import matplotlib.pyplot as plt
from keras.preprocessing import image
import numpy as np
from keras.applications.vgg16 import preprocess_input, decode_predictions
from keras import models
tf.compat.v1.disable_eager_execution()
model = VGG16(weights='imagenet')
```

```python
img_path = '/Users/markc/Downloads/Animals.jpg'

img = tf.keras.utils.load_img(img_path, target_size=(224, 224))
img_tensor = tf.keras.utils.img_to_array(img)
img_tensor = np.expand_dims(img_tensor, axis=0)
img_tensor /= 255.

print(img_tensor.shape)

import matplotlib.pyplot as plt
plt.imshow(img_tensor[0])
plt.show()
```

```python
img_path = '/Users/markc/Downloads/Animals.jpg'
img = tf.keras.utils.load_img(img_path, target_size=(224, 224)) 
x = tf.keras.utils.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

```python
preds = model.predict(x)
print('Predicted:', decode_predictions(preds, top=3)[0])
```

```python
from keras.applications.vgg19 import VGG19
model2 = VGG19(weights='imagenet')
preds2 = model2.predict(x)
print('Predicted:', decode_predictions(preds2, top=3)[0])
```

```python
from tensorflow.keras.applications.resnet50 import ResNet50
model3 = ResNet50(weights='imagenet')
preds3 = model3.predict(x)
print('Predicted:', decode_predictions(preds3, top=3)[0])
```
